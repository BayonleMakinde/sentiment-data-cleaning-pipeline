**Data Cleaning in Pandas**


```python
import pandas as pd
```

```python
# Load the data
df = pd.read_csv(r'C:\Users\Hp\Desktop\3) Sentiment dataset.csv')
```

```python
print(df)
```

```python
# Check for nulls
df.isnull().sum()
```

```python
# Drop rows with nulls
df.dropna(inplace=True)
```

```python
# Find duplicates
df.duplicated().sum()
```

```python
#Remove duplicates
df = df.drop_duplicates()
```

```python
# Cleaning String Data (Whitespace & Case)
df['Sentiment'] = df['Sentiment'].str.strip()
```

```python
# Change to lowercase
df['Text'] = df['Text'].str.lower()
```

```python
# Dropping Unnecessary Columns
df = df.drop(columns=['Unnamed: 0'])
```

```python
# Convert Timestamp to Datetime 
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Now you can extract info:
df['Hour'] = df['Timestamp'].dt.hour
```

```python
# Show results
print("Cleaning Complete!")
print(df.info())
df.head()
```

```python
# Save your cleaned DataFrame to a new CSV file
df.to_csv('cleaned_sentiment_data.csv', index=False)
```
